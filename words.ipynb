```python
from random import randint

shtraf = (
    '''
    ---------
    |
    |
    |
    |
    |
    |
    |
    ''',
    '''
    ---------
    |      |
    |      O
    |
    |
    |
    |
    |
    ''',
    '''
    ---------
    |      |
    |      O
    |      -
    |     /|\\
    |      |
    |
    |
    ''',
    '''
    ---------
    |      |
    |      O
    |      -
    |     /|\\
    |      |
    |     / \\
    |
    ------
    '''
    )

words = ('пасха', 'машина', 'транскрипция', 'информация')
questions = ('религиозный праздник, определяющий возрождение', \
             'транспортное средство', 'передача на письме звуков', \
             'оружие 21 века')


k = randint(0, len(words)-1)
word = words[k]
question = questions[k]

##word = 'пасха'
##question = 'религиозный праздник, определяющий возрождение'

print('Угадайте слово: ', question)
print('В загаданном слове', len(word), 'букв')

result = '_' * len(word)
named = ''
sh = 0
count = 0
while result != word and sh < len(shtraf):
    print(' '.join(result))
    print('Вы называли символы: '+ ' '.join(named) if  len(named)>0 else '')
    inp = input('Введите символ/слово: ')
    if len(inp) == 1:
##        символ
        if inp in word:
            count += 1
            if inp not in named:
                named += inp
                r = ''
                for i in range(len(word)):
                    if word[i] == inp:
                        r += inp
                    else:
                        r += result[i]
                result = r
            else:
                print('Будьте внимательны! Эта буква уже была названа.')
        else:
            print('Увы - буквы нет.')
            print(shtraf[sh])
            named += inp
            sh += 1
    else:
##        слово
        if inp == word:
            result = word
            count += 1
        else:
            print('Cлово неправильное. Подумайте ещё.')
            print(shtraf[sh])
            sh += 1

print('Количество попыток угадать: ', count)
if sh < len(shtraf):
    print('Отлично! Вы лучший угадыватель года!')
    
else:
    print('Не переживай. Больше тренируйся! В следующий раз всё получится!')

print('Заходи поиграть ещё!')
```

```python
import numpy as np
import re

from tensorflow.keras.layers import Dense, LSTM, Input, Dropout, Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
from tensorflow.keras.preprocessing.sequence import pad_sequences

maxWordsCount = 1000
tokenizer = Tokenizer(num_words=maxWordsCount, filters='!–"—#$%&amp;()*+,-./:;<=>?@[\\]^_`{|}~\t\n\r«»', lower=True, split=' ', char_level=False)
tokenizer.fit_on_texts(x)

```

```python
max_text_len = 10
data = tokenizer.texts_to_sequences(x)
data_pad = pad_sequences(data, maxlen=max_text_len)
```

```python

x = data_pad
```

```python
model = Sequential()
model.add(Embedding(maxWordsCount, 128, input_length = max_text_len))
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64))
model.add(Dense(3, activation='softmax'))
model.summary()

model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=Adam(0.0001))
```

```python
history = model.fit(x, y, batch_size=1, epochs=200)
```
