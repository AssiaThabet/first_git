как найти часные производные? метод - Back propagation

Метод обратного распространения ошибки (англ. backpropagation) — метод вычисления градиента, который используется при обновлении весов многослойного перцептрона. 

ЧТобы было более понятно **зачем вообще нужен этот метод**, скажу, что чтобы модель совсем хорошо что-либо предсказывала, нужно ***подобрать алоритм***. Этот метод позволяет выбрать тот алгоритм, который хорошо себя зарекомендовал в прошлом. 

В общем, мы высчитываем определённую ошику каждого слоя и корректируем веса

Представим такую нейро-сеть, которой мы каким-то образом *подобрали веса*, у которо акт. функция - Relu и результат выходного слоя с данными 1 и 2 должен примерно являтся **0,5**. веса даны, и если мы вычислим, то получится число 3,3. 

**Вычислим ошибку** 0,5-3,3 = -2,8

Нужно найти ошибку для каждого нейрона, чтобы потом подобрать веса. Такая ошибка вычисляется следующим образом:
$$ \delta * w_i * act()'$$
где act()' - производная активационной функции. В нашем примере она - функция ***ReLu, производная которой - 1.***

в каждом последующем нейроне, $\delta^l = \delta^{l+1} * w_i * act()'$ 

 А теперь, нам нужно обновить веса связей. 

 Нам нужно ту ошибку, которая прийшла по весу умножить на скорость обучения.
 скорость обучения - 0,01

Теперь нашу **ошибку нужно умножить на вес связи, на входной нейрон и на скорость обучения(LR)**.

```python
game_of_lists = [[[473], [[536], [809]], [933]], [24], [[963], [707]], [718]]

flatten_list=[]

for i in game_of_lists:
    if type(i) == list:
        for i in i:
            n = i
            if type(n)!=int:
                while type(n)!=int:
                    for i in i:
                        n = i
                        if type(n)==list:
                            flatten_list.append(n[0])
                        else:
                            if n not in flatten_list:
                                flatten_list.append(n)
            else:
               if n not in flatten_list:
                  flatten_list.append(n)
    # else:
    #     flatten_list.append(i)
```
