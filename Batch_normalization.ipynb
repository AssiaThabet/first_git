###Batch_normalization - Это метод для ускорения обучения

Существует ковариационный сдвиг - это когда модель на разных примерах имеет разное мат ожидание и разную дисперсию, тоесть разрос относительно ожидания. 

Пример такого внешнего сдвига:
К примеру, мы обучили модель на чёрных машинах, тогда на других, её результат будет плохим. В таком случае мы просто в обучающую выборку добавляем соответствующее коливество разноцветных машин

Но есть **внутренний ковариационный сдвиг**, это когда статистическое рапределение меняется от слоя, к слою.

По сути, ничего страшного это не несёт, но вот наш пример с машинами. С каждым слоем, нейронная сеть подстраивается к новым примерам. Нам на вход поступает то машина чёрного цвета, то машина другого цвета. Тоесть весовые коофиценты только адаптировались к одним авто, как поступают другие.

Скорость обучения падает, а шаг градиетнтного спуска приходится менять.

Поэтому нам нужно данные нормализовывать, стандартизировать.

В данном алгоритме главные характеристики (мат ожидание и дисперсию) приводят к значениям 0 и 1

Такую нармолизацию советуют проводить до функции активации.

Ну в общем, есть определённая формула для её определения.

Вверху изображены формула мат ожидания и дисперсии. А слева сама функция нормализации. 
Также, из-за того, что нейронная сеть может реагировать на чёрные машины также, как и на цветные нейроная сеть теряет нелинейность.Поэтому, дополнительно с помоющи градиентного спуска, смещаются эти величины (z1, z2...)

#### Как пременять этот алгоритм в keras?
Сначала, импортируем эго в модуль


```
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization
```
Далее, в создании модели, после того слоя, где он пременяется пишем BatchNormalization()



```
model = keras.Sequential([
    Flatten(input_shape=(28, 28, 1)),   
    Dense(128, activation='relu'), 
    BatchNormalization(),
    Dense(10, activation='softmax') ])
```
