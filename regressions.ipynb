## Линейная регрессия

В лнейной модели всё не сложно, у нас есть функция $ y = w_1 x + w_0$, отклонение истинного от предсказанного значения которого мы хотим минимизировать. Задача - найти веса функции. Минимизация может происходить разными методами.

Нахождение минимума в функции осуществлятется с помощью производной т.к. производная - скорость изменения функции. В формуле содержится модуль, поэтому она не дефференцируема. Берём формулу и возводим выражения в квадрат.  Это и есть ***Метод Наименьших Квадратов***. Этот метод мы будет пременять во всех регрессионных моделях, кроме нейронных сетей и логистической регрессии. 

Находим частные производные и состовляем систему уравнений с 2-мя неизвестными, которые являются нашими весами.

Пример:

```python
import pandas as pd
import matplotlib.pyplot as plt


arr1 = [-1,0,1,2]
arr2 = [1,0,1,4]
df = pd.DataFrame(data=[arr1,arr2]).T.rename(columns={0:'X', 1:'Y'})
```

```python
X = df.iloc[:, :-1].values 
y = df.iloc[:, 1].values

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# разделяет 80% данных на обучающий набор, а 20% данных – на набор тестов. 
#Переменная test_size -–это то место, где мы фактически указываем пропорцию набора тестов.

from sklearn.linear_model import LinearRegression # импорт инейной модели
regressor = LinearRegression() # создание модели машинного обучения
regressor.fit(X_train, y_train) # тренировка модели с данными


import numpy as np
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
df.plot(x='X', y='Y', style='o') 
abline(regressor.coef_, regressor.intercept_)
```

Всё это было сделанно с помощью библиотеки skliern, в которую всё это уже встроено. Поэтому формула ниже

$(w_1,w_0)=(w_1x_1+w_0-y_1)^2+(w_1x_2 + w_0-y_2)^2$

```python
import pandas as pd
import matplotlib.pyplot as plt


arr1 = [1,2,3,4]
arr2 = [30,7,8,1]
df = pd.DataFrame(data=[arr1,arr2]).T.rename(columns={0:'X', 1:'Y'})

df.plot(x='X', y='Y', style='o') 
plt.title('x vs y') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show()
''
```
