### Что такое градиентный спуск?

Это алгоритм, применяемый для минимизации функции.
Необходимо минимизировать функцию вида $$ y = (ϴ - 5)^2$$
Минимизировать значит найти при каком значении $ϴ$ функция $f(ϴ)$ приймет минимальное значение

bandicam 2022-03-13 17-54-38-339.jpg

```python
STEP_COUNT = 25
STEP_SIZE = 0.1  # Скорость обучения


def func(x):
    return (x - 5) ** 2


def func_derivative(x):
    return 2 * (x - 5)

previous_x, current_x = 0, 0

for i in range(STEP_COUNT):
    current_x = previous_x - STEP_SIZE * func_derivative(previous_x)
    previous_x = current_x

print("After", STEP_COUNT, "steps theta=", format(current_x, ".6f"), "function value=", format(func(current_x), ".6f"))
```
